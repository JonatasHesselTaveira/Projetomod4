```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
jun = pd.read_csv("https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/shpc/dsan/2021/2021-06-gasolina-etanol.csv", delimiter = ";")
jun.head()
```

```python
jun['Mes'] = 'junho'
```

```python
# Carregando os dados do mês de julho com informações fornecidas pela ANP
jul = pd.read_csv("https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/shpc/dsan/2021/2021-07-gasolina-etanol.csv", delimiter = ";")
jul.head()
```

```python
jul['Mes'] = 'julho'
```

```python
df = pd.concat([jun, jul], ignore_index=True)
```

```python
df.head()
```

```python
df.columns
```

```python
dflimpo = df[['Região - Sigla', 'Município', 'Produto', 'Valor de Venda']]
dflimpo['Valor de Venda'] = dflimpo['Valor de Venda'].replace(',','.',regex=True).astype(float)
dflimpo
```

##Questão 5:

```python
mediamunicipio = dflimpo.groupby(['Produto','Município']).mean().sort_values(by='Valor de Venda', ascending=True).reset_index()
mediamunicipio[mediamunicipio['Produto']=='ETANOL'].head(1)
```

```python
mediamunicipio[mediamunicipio['Produto']=='GASOLINA'].head(1)
```

```python
mediamunicipio[mediamunicipio['Produto']=='GASOLINA ADITIVADA'].head(1)
```
